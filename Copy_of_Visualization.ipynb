```
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
high_temp = [32, 35, 45, 55, 65, 75]

fig, ax = plt.subplots()
ax.plot(months, high_temp)
plt.show()
```

```
fig, ax = plt.subplots()  #empty Plots
ax.plot(months, high_temp)
ax.set_xlabel("Months")
ax.set_ylabel("Temperature")
ax.set_title("Months High Temprature")
plt.show()
```

```
low_temp = [20, 22, 30, 40, 50, 60]
fig, ax = plt.subplots()  #empty Plots
ax.plot(months, low_temp)
ax.set_xlabel("Months")
ax.set_ylabel("Temperature")
ax.set_title("Months High Temprature")
plt.show()
```

Coco cola sales Analysis

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

df = pd.read_csv(r'coca_cola_sales_data.csv') # , sep='\t'
df.head()

```

```
# Sales Over time

df['Invoice Data'] = pd.to_datetime(df['Invoice Data'])
fig, ax = plt.subplots()
ax.plot(df['Invoice Data'],df['Total Sales'])
ax.set_xlabel("Invoice Data")
ax.set_ylabel("Sales")
ax.set_title("Sales Over time")
```

```
df.info()
```

```
# Total Sales in Object type Code conver into integer

for col in ["Total Sales"] :
  df[col] = df[col].replace(r'[\$,]', '', regex=True).astype(int)
  print(col)
```

```
# Monthwise data
df["Year-Month"] = df["Invoice Data"].dt.to_period("M")
monthly_sales = df.groupby("Year-Month")["Total Sales"].sum()
print(monthly_sales)
```

```

fig, ax = plt.subplots()
monthly_sales.plot(marker='o')
plt.title('Coca Cola Montly Seles')
plt.xlabel('Month')
plt.ylabel('Total Sales')
ax.legend()
ax.grid()
plt.show()
```

Which montly contribute the highest and lowest total sales?

```
monthly_sales.plot(kind='bar')
plt.grid()
```

```
# complate the correlation matrix

correlation_matrix = df[['Total Sales']].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f",linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
