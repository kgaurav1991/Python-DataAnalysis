```python
# Quantile Regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
squaringFucntion = PolynomialFeatures(degree=2)
Pesticide_square = squaringFucntion.fit_transform(df[['Pesticide_Amount']])
Pesticide_square
```

```python
parabola = LinearRegression()
parabola.fit(Pesticide_square,df['Plant_Production'])
```

```python
plt.scatter(df['Pesticide_Amount'], df['Plant_Production'],color='red',marker='+')
plt.plot(df['Pesticide_Amount'],parabola.predict(Pesticide_square),color='gray')
plt.xlabel('Pesticide Amount')
plt.ylabel('Plant Production')
plt.title('Plant Yield Prediction')
plt.show()
```

```python
# Your Tasks:
#1. check for missing values.
#2. fill missing Test_Score with the Mean
#3. fill missing Experience_Yrs with the Median
#4. Perform lable Encoding on the Degree colums manually using a map
#5. Create a new feature called Score_Per_Year(Test_Score/Experience_Yrs).
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

hr_data = {
    'Candidate' : ['A','B','C','D','E'],
    'Test_Score' : [85,90,np.nan,75,88],
    'Experience_Yrs' : [2,5,3,np.nan,4],
    'Degree' : ['Bachelor','Master','PhD','Bachelor','Master']
}

```

```python
#1. check for missing values.

df_hr = pd.DataFrame(hr_data)
df_hr
```

```python
#1. check for missing values.
df_hr.isnull().sum()
```

```python
#2. fill missing Test_Score with the Mean
df_hr['Test_Score'] = df_hr['Test_Score'].fillna(df_hr['Test_Score'].mean())
df_hr
```

```python
#3. fill missing Experience_Yrs with the Median
df_hr['Experience_Yrs'] = df_hr['Experience_Yrs'].fillna(df_hr['Experience_Yrs'].median())
df_hr
```

```python
#4. Perform lable Encoding on the Degree colums manually using a map
degree_map = {
    'Bachelor' : 1,
    'Master' : 2,
    'PhD' : 3
}
df_hr['Degree'] = df_hr['Degree'].map(degree_map)
df_hr
```

```python
#5. Create a new feature called Score_Per_Year(Test_Score/Experience_Yrs).
df_hr['Score_Per_Year'] = df_hr['Test_Score'] / df_hr['Experience_Yrs']
df_hr
```
