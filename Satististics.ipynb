IQR Method to Remove outlier

```python
Q1 = df['Monthly Income'].quantile(0.25)
Q3 = df['Monthly Income'].quantile(0.75)
IQR = Q3 - Q1
IQR
# Upper Range = Q3 + 1.5*IQR
upper_range = Q3 + 1.5*IQR

outliers = df[
    (df['Monthly Income'] < upper_range)
]

outliers
```

```python
Q1 = df['Monthly Income'].quantile(0.25)
Q3 = df['Monthly Income'].quantile(0.75)
IQR = Q3 - Q1
IQR
# Lower Range = Q3 - 1.5*IQR
lower_range = Q3 - 1.5*IQR

outliers = df[
    (df['Monthly Income'] > lower_range)
]

outliers
```

Both Outliers Remove

```python
Q1 = df['Monthly Income'].quantile(0.25)
Q3 = df['Monthly Income'].quantile(0.75)
IQR = Q3 - Q1
IQR
# Upper Range = Q3 + 1.5*IQR
upper_range = Q3 + 1.5*IQR
lower_range = Q3 - 1.5*IQR

outliers = df[
    (df['Monthly Income'] < upper_range) &
    (df['Monthly Income'] > lower_range)
]

outliers
```

```python
# Z score
# upper_range = mean + 3*(sigma)
import pandas as pd
import numpy as np

df = pd.read_csv('income.csv')
df.head()

```

```python
df.describe()
```

```python
# Z score
# upper_range = mean + 3*(sigma)

mean = df['Monthly Income'].mean()
sigma = df['Monthly Income'].std()

upper_range = mean + 3*(sigma)
upper_range
```

```python
df = df[df['Monthly Income'] < upper_range]
df
```

```python
# Z score
# upper_range = mean + 2*(sigma)

mean = df['Monthly Income'].mean()
sigma = df['Monthly Income'].std()

upper_range = mean + 2*(sigma)
upper_range
```

```python
df = df[df['Monthly Income'] < upper_range]
df
```
